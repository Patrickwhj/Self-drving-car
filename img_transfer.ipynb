```python
# coding=utf-8
# Transfer format of images from jpg to npz
import os
import numpy as np
from time import time
import math
from PIL import Image
from sklearn.preprocessing import OneHotEncoder
import img_global

CHUNK_SIZE = 128    # Dividing the images into chunks

def img_mapping(img_path, key):
    print(img_path, key)
    image = Image.open(img_path)
    img_array = np.array(image)
    img_array = np.expand_dims(img_array, axis=0)  # add one dimension
    
    # one-hot encoding
    if key == 1:
        label_array = [ 1., 0., 0.]
    elif key == 2:
        label_array = [ 0., 1., 0.]
    elif key == 3:
        label_array = [ 0., 0., 1.]
    

    return (img_array, label_array)


if __name__ == '__main__':
    path = './img_data'
    files = os.listdir(path)                             
    turns = int(math.ceil(len(files) / CHUNK_SIZE))      
    print("number of files: {}".format(len(files)))
    print("turns: {}".format(turns))

    for turn in range(0, turns):

        img_labels = np.zeros((1, OUTPUT_SHAPE ), 'float')           
        img_data = np.zeros([1, HEIGHT, WIDTH, CHANNELS])            

        CHUNK_files = files[turn * CHUNK_SIZE: (turn + 1) * CHUNK_SIZE] 
        print("number of CHUNK files: {}".format(len(CHUNK_files)))
        for file in CHUNK_files:
            # decide the format
            if not os.path.isdir(file) and file[len(file) - 3:len(file)] == 'jpg':
                try:
                    key = int(file[0])                     
                    img_array, label_array = img_mapping(path + "/" + file, key)
                    img_data = np.vstack((img_data, img_array))
                    img_labels = np.vstack((img_labels, label_array))
                except:
                    print('process error')

        
        img_data = img_data[1:, :]
        img_labels = img_labels[1:, :]
        file_name = str(time())
        directory = "./img_data_npz"

        if not os.path.exists(directory):
            os.makedirs(directory)
        try:
            np.savez(directory + '/' + file_name + '.npz', img_data=img_data, img_labels=img_labels)
            print('Current turn!!'+str(turn))
        except IOError as e:
            print(e)
```
